```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)  
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['keywords'] = movies['keywords'].apply(convert)  
```

```python
movies['cast'][0]
```

```python
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3) 
```

```python
movies.head() 
```

```python
movies['crew'][0]
```

```python
# I want only that column where 'job = Director'
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L    
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

# Text Vectorization
various techniques for text vectorization:
1. Bag of words
2. tfidf
3. word2vec

Here, I use 1st one - bag of words
step 1 -- first concatenate all the tags -- tag1 + tag2 + ....
step 2 -- it becomes a large pool of words, and I need 5000 most common words like W1, W2, W3, ....
step 3 -- then suppose 
      for movie 1 -->        
              w1 = action -- it's frequency = 5 times
          and w2 = future -- it's frequency = 3 times
          and w3 freq = 0
          .... w5000 freq = 3
      for movie 2 -->        
              w1 = action -- it's frequency = 2 times
          and w2 = future -- it's frequency = 0 times
          and w3 freq = 1
           ......w5000 freq = 2

           ......repeat for all movies

comes a table of 5000 x 5000 order 'number of movies' and 'number of most common words'

* In vectorization, we don't use "stop words" like are, of , to , from etc. which are used in sentence formation.
* we use "sk-learn library" for text vectorization i.e. 'countvectorizer'

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
# using nltk library
import nltk 
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

# Now, I will create a 'helper_function'

```python
def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
    
    return " ".join(y)
    
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

# we will calculate the 'cosine distance' between two vectors, not the 'euclidean distance'
* higher distance --> less similarity and vice-versa
* for 'high dimensional data' -->  'euclidean distance' formula fails.
* for this problem --> I have to use 'cosine_similarity'

# cosine_similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors)
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity
```

```python
similarity[2]
```

```python
# sorting

sorted(similarity[0], reverse=True)
```

```python
# I will call a "enumerate_function" --> to convert data into list of tuples
# to calculate nearest 05 movies list

sorted(list(enumerate(similarity[0])), reverse = True, key=lambda x:x[1])[1:6]
```

# We will make a 'Recommend_Function' and in return it will give '05_similar_movies' name

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse = True, key=lambda x:x[1])[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]].title)
    
```

```python
# new_df[new_df['title'] == 'Batman Begins'].index[0]
```

```python
recommend('Batman Begins')
```
